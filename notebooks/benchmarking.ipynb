```julia
using BenchmarkTools

# Create a fake ϕ map
ϕ_plain = rand(100, 100)
ϕ_cartesian = copy(ϕ_plain)

# (1) Plain i, j loops
function update_plain!(ϕ)
    for i in 2:99
        for j in 2:99
            ϕ[i,j] += 0.001
        end
    end
end

# (2) CartesianIndices
function update_cartesian!(ϕ)
    for idx in CartesianIndices((2:99, 2:99))
        ϕ[idx] += 0.001
    end
end

# Benchmark
println("Benchmarking plain i,j loops:")
@btime update_plain!($ϕ_plain)

println("Benchmarking CartesianIndices:")
@btime update_cartesian!($ϕ_cartesian)
```
